```python
#Q3-  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x: int, the outcome (0 or 1)
    - p: float, probability of success (0 â‰¤ p â‰¤ 1)

    Returns:
    - float, the probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print(bernoulli_pdf(1, 0.7))
print(bernoulli_pdf(0, 0.7))
```

```python
#Q9- Write a Python function to calculate Z-scores from a dataset and plot them?
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for the given dataset.

    Parameters:
    - data: list or numpy array, the dataset

    Returns:
    - z_scores: numpy array, Z-scores corresponding to the data points
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 1000)  # 1000 samples from a normal distribution

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 5))
plt.plot(z_scores, label='Z-scores', color='b')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.axhline(y=0, color='r', linestyle='--', label='Mean (Z = 0)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#10-Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_param = 1    # rate parameter for the exponential distribution
sample_sizes = [5, 30, 100]  # Different sample sizes to observe CLT
num_samples = 10000  # Number of sampling iterations

# Function to simulate sampling from the Exponential distribution and calculate sample means
def simulate_clt(sample_size, num_samples, lambda_param):
    # Sample from Exponential distribution
    samples = np.random.exponential(scale=1/lambda_param, size=(num_samples, sample_size))

    # Calculate the sample means
    sample_means = np.mean(samples, axis=1)

    return sample_means

# Plotting the results
plt.figure(figsize=(12, 8))

for i, sample_size in enumerate(sample_sizes, start=1):
    sample_means = simulate_clt(sample_size, num_samples, lambda_param)

    # Plotting the histogram of sample means
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, edgecolor='black')
    plt.title(f'Sample Size = {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

# Add a title for the whole plot
plt.suptitle(f'Central Limit Theorem for Exponential Distribution (Î»={lambda_param})', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
#Q14- Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, data):
    """
    Calculates the Z-score for a given data point x in a dataset.

    Parameters:
    - x: float, the data point
    - data: list or numpy array, the dataset

    Returns:
    - z: float, the Z-score of x
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z = (x - mean) / std_dev
    return z

# Example dataset (normally distributed)
data = np.random.normal(loc=100, scale=15, size=1000)

# Data point to analyze
x_point = 130

# Calculate Z-score
z = calculate_z_score(x_point, data)
print(f"Z-score for x = {x_point}: {z:.2f}")

# Plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, 0, y, where=(x <= z), color='skyblue', alpha=0.5, label='Area â‰¤ Z')

# Mark the Z-score
plt.axvline(x=z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.title('Z-score Visualization on Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q15- Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from scipy.stats import norm

def perform_z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for population mean.

    Parameters:
    - sample_data: list or array of sample observations
    - pop_mean: hypothesized population mean (Î¼â‚€)
    - pop_std: known population standard deviation (Ïƒ)
    - alpha: significance level (default = 0.05)

    Returns:
    - z_stat: computed Z-statistic
    - p_value: two-tailed p-value
    - conclusion: hypothesis test result
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    se = pop_std / np.sqrt(n)

    # Z-statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Hypothesis decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis"
    else:
        conclusion = "Fail to reject the null hypothesis"

    return z_stat, p_value, conclusion

# --- Example ---
# Population mean (Î¼â‚€)
mu_0 = 100

# Known population std dev (Ïƒ)
sigma = 15

# Sample data (simulated)
np.random.seed(0)
sample = np.random.normal(loc=103, scale=15, size=50)

# Perform Z-test
z_stat, p_value, conclusion = perform_z_test(sample, mu_0, sigma)

# Display results
print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")

```

```python
#Q16-Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data: list or numpy array of sample values
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - mean: sample mean
    - lower_bound, upper_bound: confidence interval bounds
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2, df=n - 1)

    lower_bound = mean - margin
    upper_bound = mean + margin
    return mean, lower_bound, upper_bound

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Calculate 95% confidence interval
mean, ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

# Display the result
print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q3- Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05, tail='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample_data (array-like): Sample data points.
        population_mean (float): Known population mean.
        population_std (float): Known population standard deviation.
        alpha (float): Significance level (default is 0.05).
        tail (str): 'two-sided', 'left', or 'right' (for one-tailed tests).

    Returns:
        z_score (float): Computed Z-score.
        p_value (float): Corresponding p-value.
        result (str): Interpretation of the hypothesis test.
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Z-score calculation
    z_score = (sample_mean - population_mean) / standard_error

    # P-value calculation based on test type
    if tail == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif tail == 'left':
        p_value = norm.cdf(z_score)
    elif tail == 'right':
        p_value = 1 - norm.cdf(z_score)
    else:
        raise ValueError("tail must be 'two-sided', 'left', or 'right'")

    # Decision
    if p_value < alpha:
        result = "Reject the null hypothesis."
    else:
        result = "Fail to reject the null hypothesis."

    return z_score, p_value, result

# --- Example Usage ---
np.random.seed(42)
sample = np.random.normal(loc=52, scale=10, size=30)  # Sample drawn from mean 52
pop_mean = 50
pop_std = 10

z, p, interpretation = one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05, tail='two-sided')

print(f"Z-Score: {z:.4f}")
print(f"P-Value: {p:.4f}")
print(f"Conclusion: {interpretation}")
```

```python
#Q5-Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, tail='right'):
    """
    Visualize Type I and Type II errors for hypothesis testing.

    Parameters:
        mu0 (float): Mean under null hypothesis (H0)
        mu1 (float): Mean under alternative hypothesis (H1)
        sigma (float): Population standard deviation
        n (int): Sample size
        alpha (float): Significance level
        tail (str): 'right', 'left', or 'two-sided'
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for H0
    if tail == 'right':
        z_critical = norm.ppf(1 - alpha)
        x_crit = mu0 + z_critical * se
    elif tail == 'left':
        z_critical = norm.ppf(alpha)
        x_crit = mu0 + z_critical * se
    elif tail == 'two-sided':
        z_critical = norm.ppf(1 - alpha / 2)
        x_crit_low = mu0 - z_critical * se
        x_crit_high = mu0 + z_critical * se
    else:
        raise ValueError("tail must be 'right', 'left', or 'two-sided'")

    # X range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='Hâ‚€ Distribution', color='blue')
    plt.plot(x, y_alt, label='Hâ‚ Distribution', color='green')

    if tail == 'right':
        # Type I Error (alpha)
        plt.fill_between(x, y_null, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (Î±)')
        # Type II Error (beta)
        plt.fill_between(x, y_alt, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (Î²)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    elif tail == 'left':
        plt.fill_between(x, y_null, where=(x <= x_crit), color='red', alpha=0.4, label='Type I Error (Î±)')
        plt.fill_between(x, y_alt, where=(x > x_crit), color='orange', alpha=0.4, label='Type II Error (Î²)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    elif tail == 'two-sided':
        plt.fill_between(x, y_null, where=(x <= x_crit_low) | (x >= x_crit_high), color='red', alpha=0.4, label='Type I Error (Î±)')
        plt.fill_between(x, y_alt, where=(x > x_crit_low) & (x < x_crit_high), color='orange', alpha=0.4, label='Type II Error (Î²)')
        plt.axvline(x_crit_low, color='black', linestyle='--', label=f'Critical Low = {x_crit_low:.2f}')
        plt.axvline(x_crit_high, color='black', linestyle='--', label=f'Critical High = {x_crit_high:.2f}')

    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# --- Example Usage ---
visualize_type1_type2_errors(
    mu0=100,     # H0 mean
    mu1=105,     # H1 mean
    sigma=15,    # known population std dev
    n=36,        # sample size
    alpha=0.05,  # significance level
    tail='right' # one-tailed test (right)
)
```

```python
#Q9-  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        sample (array-like): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound): Confidence interval bounds
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    margin_of_error = t_critical * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error
    return (lower, upper)

# --- Example usage ---
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")

```

```python
#Q11- Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayesian_inference(P_H, P_E_given_H, P_E_given_not_H, P_not_H):
    """
    Perform Bayesian Inference using Bayes' Theorem.

    Parameters:
        P_H (float): Prior probability of hypothesis (e.g., having the disease)
        P_E_given_H (float): Likelihood (probability of evidence given hypothesis)
        P_E_given_not_H (float): Likelihood (probability of evidence given not hypothesis)
        P_not_H (float): Prior probability of the negation of hypothesis (e.g., not having the disease)

    Returns:
        float: Posterior probability of hypothesis given evidence
    """
    # Calculate the evidence P(E)
    P_E = P_E_given_H * P_H + P_E_given_not_H * P_not_H

    # Calculate the posterior probability P(H|E)
    P_H_given_E = (P_E_given_H * P_H) / P_E
    return P_H_given_E

# --- Example usage ---
# Prior probabilities and likelihoods
P_H = 0.1  # Probability of having the disease (prior)
P_E_given_H = 0.95  # Probability of testing positive given the disease
P_E_given_not_H = 0.05  # Probability of testing positive given no disease
P_not_H = 1 - P_H  # Probability of not having the disease

# Perform Bayesian inference
posterior = bayesian_inference(P_H, P_E_given_H, P_E_given_not_H, P_not_H)

print(f"Posterior Probability of having the disease given a positive test result: {posterior:.4f}")
```

```python
#Q12- Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (cross-tabulation)
# Example: Relationship between Gender (Male, Female) and Preference (A, B)
data = {
    'Preference A': [30, 10],  # 30 males prefer A, 10 females prefer A
    'Preference B': [20, 40]   # 20 males prefer B, 40 females prefer B
}

# Convert the data to a DataFrame
df = pd.DataFrame(data, index=['Male', 'Female'])
print("Contingency Table:")
print(df)

# Step 2: Perform Chi-square test
chi2, p_value, dof, expected = chi2_contingency(df)

# Step 3: Output results
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")

# Step 4: Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is an association between gender and preference.")
else:
    print("\nFail to reject the null hypothesis: There is no association between gender and preference.")
```

```python
#Q13- Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
        observed (ndarray): 2D array of observed frequencies.

    Returns:
        ndarray: 2D array of expected frequencies.
    """
    # Step 1: Calculate row totals, column totals, and grand total
    row_totals = np.sum(observed, axis=1)  # Sum of each row
    col_totals = np.sum(observed, axis=0)  # Sum of each column
    grand_total = np.sum(observed)         # Total sum of all values

    # Step 2: Calculate expected frequencies using the formula
    expected = np.outer(row_totals, col_totals) / grand_total

    return expected

# --- Example usage ---
# Example contingency table: Relationship between Gender (Male, Female) and Preference (A, B)
observed_data = np.array([[30, 20],  # 30 males prefer A, 20 males prefer B
                          [10, 40]]) # 10 females prefer A, 40 females prefer B

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#Q14- Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):
    """
    Perform a Chi-square goodness-of-fit test.

    Parameters:
        observed (array-like): Observed frequencies.
        expected (array-like): Expected frequencies (under the null hypothesis).

    Returns:
        chi2_stat (float): Chi-square statistic.
        p_value (float): p-value for the test.
    """
    chi2_stat, p_value = chisquare(observed, expected)
    return chi2_stat, p_value

# --- Example usage ---
# Observed frequencies (e.g., outcome of 60 rolls of a fair die)
observed_data = np.array([10, 12, 9, 11, 8, 10])  # Rolls of 1, 2, 3, 4, 5, 6

# Expected frequencies (assuming a fair die with equal probability for each face)
expected_data = np.array([10, 10, 10, 10, 10, 10])  # Uniform distribution: 1/6 probability for each face

# Perform the goodness-of-fit test
chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 3: Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution.")
```
